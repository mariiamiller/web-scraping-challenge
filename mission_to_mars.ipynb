```python
url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
```

```python
html=browser.html
soup = bs(html, 'html.parser')
results = soup.find_all(class_="item")
results
```

```python

photo_dict = {}
hemisphere_image_urls = []
for result in results:
    photo_dict['title']=result.h3.text
    browser.click_link_by_partial_text(result.h3.text)
    html = browser.html
    soup = bs(html, 'html.parser')
    img_search = soup.find_all(class_='wide-image')
    photo_dict['img_url']='https://astrogeology.usgs.gov'+img_search[0]['src']
    print(photo_dict)
    hemisphere_image_urls.append(photo_dict)
    url='https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
    browser.visit(url)

hemisphere_image_urls
```
